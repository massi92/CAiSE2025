```python
from langchain_community.document_loaders.directory import DirectoryLoader, TextLoader
from langchain_text_splitters import CharacterTextSplitter, RecursiveCharacterTextSplitter, RecursiveJsonSplitter
from langchain_community.vectorstores import FAISS
```

```python
loader = DirectoryLoader("/content/input_files", show_progress=True, loader_cls=TextLoader)
```

```python
docs = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=20)
documents = text_splitter.split_documents(docs)

vector = FAISS.from_documents(documents, embeddings)
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import PromptTemplate
from langchain.schema import StrOutputParser

promptT1 = PromptTemplate.from_template("""
Consider the definitions (between <definitions>
</definitions>) of step and task of analysis scenarios.

<definitions>
An analysis scenario is composed of steps and each step is accomplished through a series of tasks. You can think of a step as a macro-group of affine tasks.
</definitions>

I will provide you with a textual description corresponding to an analysis scenario.
As output print the ordered list of steps between <steps></steps>.
Do not detail the tasks of each step. I will ask you about them later.

Base your answer only on the analysis scenarios provided in the given context:
<context>
{context}
</context>

Analysis scenario: {input}

""")

promptT2 = PromptTemplate.from_template("""
Consider the definitions (between <definitions>
</definitions>) of step and task of analysis scenarios.

<definitions>
An analysis scenario is composed of steps and each step is accomplished through a series of tasks. You can think of a step as a macro-group of affine tasks.
</definitions>

Given the list of steps (enclosed between <steps>
</steps>) detail the tasks to be executed to accomplish the step {step}".
Enclose the ordered list of tasks between <tasks></tasks>.

{input}

Base your answer only on the analysis scenarios provided in the given context:
<context>{context}</context>

""")

```

```python
document_chain_1 = create_stuff_documents_chain(llm, promptT1)

document_chain_2 = create_stuff_documents_chain(llm, promptT2)
```

```python
from langchain.chains import create_retrieval_chain
from langchain.schema.runnable import RunnablePassthrough

retriever = vector.as_retriever()
```

```python
scenario_desc = """
I have to read a database table, containing 10 columns: 1 column for the record ID, 1 column for the variable short name, 1 column for a timestamp value, 6 columns containing electrical current values and a column containing torque values (in Newton meter) collected from a smart machine engine.
The columns containing the record ID and the variable short name are not required for the next analysis steps.
I want to use the 6 columns containing electrical current values for building a LASSO regression model and the torque has to be the prediction target for the model.
Afterwards, I want to compare the predicted torque values with the real values.
"""
```

```python
chain_T1 = create_retrieval_chain(retriever, document_chain_1)
```

```python
response_T1 = chain_T1.invoke({"input": scenario_desc})
```

```python
print(response_T1["answer"])

obtained_steps = response_T1["answer"]
```

```python
step = """
with name="Model training (LASSO regression model) using electrical currents as features and torque as target variable"
"""
```

```python
chain_T2 = create_retrieval_chain(retriever, document_chain_2)

response_T2 = chain_T2.invoke({"input": obtained_steps, "step": step})
```
